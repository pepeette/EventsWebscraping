```python
import requests
url = 'https://www.eventsforgamers.com/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
result = requests.get(url, headers=headers)
print(result.content.decode())
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(result.content, "html.parser")
```

```python
import json
data = [
    json.loads(x.string) for x in soup.find_all("script", type="application/ld+json")
]
data[1][0]
```

```python
import pandas as pd

event = []
start_date = []
end_date = []
location = []

for d in data[1]:
    event.append(d['name'])
    start_date.append(d['startDate'])
    end_date.append(d['endDate'])
    try:
        location.append(d['location']['address']['addressCountry'])
    except Exception:
        location.append('n/a')
    
    
df = pd.DataFrame(list(zip(event,location, start_date, end_date)),
               columns =['name', 'location', 'start', 'end'])
df
```
